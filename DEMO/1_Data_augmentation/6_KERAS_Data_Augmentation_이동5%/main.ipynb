```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
        width_shift_range = 0.05,
        height_shift_range = 0.05)       
```

```python
import os
```

```python
file_list_detail = []
file_list = os.listdir()
for i in file_list:
    if 'image' in i:
        file_list_detail_temp = os.listdir(i+"/")
        # print (i,os.listdir(i+"/"))
        for j in file_list_detail_temp:
            file_list_detail.append(i+"\\"+j)  
```

```python
file_list_detail
```

```python
for i, f in enumerate(file_list_detail):   
    label01, label02 = f.split('\\')
    label03, label04 = label02.split('.')
```

```python
for i, f in enumerate(file_list_detail):  
    
    label01, label02 = f.split('\\')
    label03, label04 = label02.split('.')
    
    img = load_img(f)
    x = img_to_array(img)                              
    x = x.reshape((1,) + x.shape)                    
    a = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir='KERAS_Data_Augmentation/', save_prefix=label03+"-KERAS_Data_Augmentation", save_format='png'):
        a += 1
        #print (batch)
        if a > 0:
            break  
        
    print (i)
print ("Dpne")
```
