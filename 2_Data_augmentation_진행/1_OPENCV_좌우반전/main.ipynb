```python
file_list_detail = []
file_list = os.listdir()
for i in file_list:
    if 'image' in i:
        file_list_detail_temp = os.listdir(i+"/")
        print (i,os.listdir(i+"/"))
        for j in file_list_detail_temp:
            file_list_detail.append(i+"\\"+j)   
```

```python
for i in file_list_detail:
    #print (i)
    label1, label2 = i.split('\\')
    label3, label4 = label2.split('.')
    print(label3)
```

```python
import cv2
import numpy as np

for i, f in enumerate(file_list_detail):  
    label1, label2 = f.split('\\')
    label3, label4 = label2.split('.')
    print(i)
    
    
    img= cv2.imread(f)
    output_3 = cv2.flip(img, 1)
    cv2.imwrite(label1+"/"+label3+"-flip1.png", output_3)
    
print("Done")
```
